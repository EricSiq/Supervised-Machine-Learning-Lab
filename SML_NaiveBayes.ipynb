```python
'''
3 types of Naive Bayes Algorithm:

Gaussian NB
Multinomial NB
Bernoulli NB
'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python

heartdata = pd.read_csv('heart.csv')
heartdata.head()
```

```python


# Split the dataset into features (X) and target (y)
X = heartdata.drop('target', axis=1)
y = heartdata['target']

# Handle missing values (replace with mean for numerical features)
X = X.fillna(X.mean())

# Scale the features to a range of 0 to 1
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Create and train different Naive Bayes models
gaussian_nb = GaussianNB()
bernoulli_nb = BernoulliNB()
multinomial_nb = MultinomialNB()

# Train models
gaussian_nb.fit(X_train, y_train)
bernoulli_nb.fit(X_train, y_train)
multinomial_nb.fit(X_train, y_train)

# Make predictions
y_pred_gaussian = gaussian_nb.predict(X_test)
y_pred_bernoulli = bernoulli_nb.predict(X_test)
y_pred_multinomial = multinomial_nb.predict(X_test)

# Evaluate the models
accuracy_gaussian = accuracy_score(y_test, y_pred_gaussian)
accuracy_bernoulli = accuracy_score(y_test, y_pred_bernoulli)
accuracy_multinomial = accuracy_score(y_test, y_pred_multinomial)
print(f"Using Heart Data: ")
print(f"Gaussian Naive Bayes Accuracy: {accuracy_gaussian}")
print(f"Bernoulli Naive Bayes Accuracy: {accuracy_bernoulli}")
print(f"Multinomial Naive Bayes Accuracy: {accuracy_multinomial}")
```

```python


# Data for the bar chart
naive_bayes_models = ['Gaussian NB', 'Bernoulli NB', 'Multinomial NB']
accuracies = [accuracy_gaussian, accuracy_bernoulli, accuracy_multinomial]

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(naive_bayes_models, accuracies, color=['blue', 'green', 'orange'])
plt.title('Heart Data:Naive Bayes Model Accuracy Comparison')
plt.xlabel('Naive Bayes Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Set y-axis limit to 0-1 for accuracy
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add data labels on top of the bars
for i, v in enumerate(accuracies):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center', va='bottom')

plt.show()
```
